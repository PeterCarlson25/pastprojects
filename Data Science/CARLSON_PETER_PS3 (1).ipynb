```python
#π(k+1) = αH π(k) + [αd.π(k) + (1 −α)]a for influence vector

first_iteration = np.ones(len(H)) / len(H)

def Eigenvector_iterator(firstiter,normalized,article,dangle):
#create variable to iterate on the first one
    count = firstiter
    iterationcount = 0
#(pi at k+1) - (pi at k) is greater than epsilon
    remainingL1 = Epsilon_constant
    while remainingL1 >= Epsilon_constant:
#find the absolute value (l1 distance)  np.absolute and then sum of np.absolute
#create variable to identify current iteration
        newvector=(α*np.matmul(normalized,count))+(((α*np.matmul(dangle,count)+(1-α)))*article)
                                   
        remainingL1=(np.sum(np.absolute(newvector - count)))
#count needs to change to newvector so it changes for the next iteration        
        count=newvector
    
        iterationcount = iterationcount + 1
    print(iterationcount,"iterations")
    return newvector, iterationcount
```
